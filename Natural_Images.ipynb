# Data Preprocessing

```python
path = kagglehub.dataset_download("prasunroy/natural-images") + '/natural_images'
print('Path to the dataset: ', path)

def load_image(image_path, resize=True):
  img = cv2.cvtColor(cv2.imread(image_path), cv2.COLOR_BGR2RGB)
  if resize:
    img = cv2.resize(img, (256, 256), interpolation = cv2.INTER_AREA)
  
  return img

def show_image(img_path):
  img = load_image(img_path)
  plt.imshow(img)
  plt.axis('off')

training_folders = sorted(glob.glob(path + '/*'))
print(len(training_folders))
random_image_paths = [np.random.choice(glob.glob(f'{tf}/*.jpg')) for tf in training_folders]



```
